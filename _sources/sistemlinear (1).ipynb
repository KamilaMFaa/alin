# Sistem Persamaan Linear

Sistem persamaan linear merupakan sekumpulan persamaan linear yang memuat sejumlah variabel, koefisien, dan kostanta. Biasanya digambarkan sebagai garis lurus pada bidang datar. Bentuk umumnya :

$
\begin{array}{ccccccccccc}
a_{1,1}x_{1} & + & a_{1,2}x_{2} & + & \cdots & + & a_{1,j}x_{j} & = & b_{1} \\
a_{2,1}x_{1} & + & a_{2,2}x_{2} & + & \cdots & + & a_{2,j}x_{j} & = & b_{2} \\
\vdots & & \vdots & & \ddots & &  \vdots & &  \vdots \\
a_{i,1}x_{1} & + & a_{i,2}x_{2} & + & \cdots & + & a_{i,j}x_{j} & = & b_{i} \\
\end{array}
$

dengan:
- i adalah urutan suku, dimana suku merupakan bagian dari suatu persamaan yang terdiri dari variabel, koefisien dan konstanta.
- j adalah urutan persamaan.
- $a_{1,1}, a_{1,2}, ..., a_{ij} \in R$ adalah koefisien, dimana koefisisen merupakan bilangan yang menjabarkan jumlah variabel yang sejenis dan biasanya terletak di depan variabel.
- $x_1, x_2..., x_j$ adalah variabel, dimana variabel merupakan pengganti bilangan yang belum diketahui nilainya dan biasanya dilambangkan dengan huruf.
- $b_1, b_2, ...,b_i \in R$ adalah konstanta, dimana konstanta merupakan nilai bilangan konstan yang tidak diikuti variabel belakangnya.

Adapun bentuk matriksnya :
$
 \begin{bmatrix}
  a_{1,1} & a_{1,2} & \cdots & a_{1,j} \\
  a_{2,1} & a_{2,2} & \cdots & a_{2,j} \\
  \vdots  & \vdots  & \ddots & \vdots  \\
  a_{i,1} & a_{i,2} & \cdots & a_{i,j} 
 \end{bmatrix}
 \begin{bmatrix}
  x_1 \\
  x_2 \\
  \vdots \\
  x_j
 \end{bmatrix}
$
$=$
$
 \begin{bmatrix}
  b_1 \\
  b_2 \\
  \vdots \\
  b_i
 \end{bmatrix}
$

atau $AX = B$
dengan:
- i adalah urutan baris
- j adalah urutan kolom
- A adalah koefisien
- X adalah variabel
- B adalah konstanta

## Solusi Persamaan Linear

Suatu sistem persamaan linear belum tentu punya solusi, keberadaan solusi ini sangat tergantung dari sistem persamaan linear itu sendiri. Suatu sistem persamaan linear (SPL), dalam keterkaitanya dengan solusi, mempunyai tiga kemungkinan, yaitu :
- SPL mempunyai solusi tunggal jika dua garis persamaan dalam diagram karteius berpotongan di satu titik. Contoh pada persamaan $x_1 + 3x_2 = 5$ dan $x_1 - 2x_2 = -5$ yang memiliki solusi $x_1=-1$, $x_2=2$
- SPL mempunyai solusi tak hingga jika dua garis persamaan dalam diagram kartesius berimpit. Contoh persamaan $x_1 + 3x_2 = 5$ dan $x_1 - 2x_2 = 0$.
- SPL tidak mempunyai solusi jika dua garis persamaan dalam diagram kartesius tidak pernah berpotongan/sejajar. Contoh persamaan $x_1 + 3x_2 = 5$ dan $2x_1 - 6x_2 = 10$.

Berikut visualisasi dari ketiga kemungkinan solusi :

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5,100)

fig, ax = plt.subplots()
plt.title("SPL Satu Solusi")
ax.plot(x,(5-x)/3)
ax.plot(x,(5+x)/2)

ax.text(1,1.6,'$x_1+3x_2 = 5$')
ax.text(-3,0.5,'$x_1-2x_2 = -5$')

ax.set_xlim(-4,4)
ax.set_ylim(-2,6)
ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)

## This options specifies the ticks based the list of numbers provided.
ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')
```

```python
fig, ax = plt.subplots()
plt.title("SPL Banyak Solusi")
ax.plot(x,(5-x)/3)
ax.plot(x,(5-x)/3)

ax.text(1,1.6,'$x_1+3x_2 = 5$')
ax.text(-3,1.2,'$2x_1+6x_2 = 10$')

ax.set_xlim(-4,4)
ax.set_ylim(-2,6)
ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)

ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')
```

```python
fig, ax = plt.subplots()
plt.title("SPL Tanpa Solusi")
ax.plot(x,(5-x)/3)
ax.plot(x,-x/3)

ax.text(1,1.6,'$x_1+3x_2 = 5$')
ax.text(0.3,-1.4,'$x_1+3x_2 = 0$')

ax.set_xlim(-4,4)
ax.set_ylim(-2,6)
ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)

## This options specifies the ticks based the list of numbers provided.
ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')
```

## Penyelesaian Sistem Persamaan Linier

### Eliminasi Gauss
Metode eliminasi gauss adalah adalah metode operasi nilai dalam matriks yang mengubah dengan menukar posisi baris, menambahkan atau mengurangi satu baris dengan baris yang lain, atau mengalikan satu baris dengan skalar.

Contoh :

$8x_1 + 9x_2 + x_3 = 19$

$x_1 + 2x_2 + 3x_3 = 9$

$2x_1 + x_2 + x_3 = 5$

Bentuk Matriks :
$
 \begin{bmatrix}
  8 & 9 & 1\\
  1 & 2 & 3\\
  2 & 1 & 1
 \end{bmatrix}
 \begin{bmatrix}
  x_1\\
  x_2\\
  x_3
 \end{bmatrix}
$
$=$
$
 \begin{bmatrix}
  19\\
  9\\
  5
 \end{bmatrix}
$

**Numerical Phyton (NumPy)**
dikembangkan pada tahun 2015 oleh Travis Oliphant, NumPy adalah library python untuk perhitungan matematis dan ilmiah. Library ini memuat fungsi aljabar linier, fourier transform, dan komputasi matriks.

Berikut bentuk matriks menggunakan NumPy :

```python
import numpy as np
A=np.array([[8,9,1,19],[1,2,3,9],[2,1,1,5]])
print(A)
```

Berikut adalah langkah-langkah Eliminasi Gauss menggunakan NumPy :

Pertama, masukkan kode NumPy yang akan digunakan nantinya seperti di bawah ini.

```python
def RowSwap(A,k,l):
# =============================================================================
#     A is a NumPy array.  RowSwap will return duplicate array with rows
#     k and l swapped.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        temp = B[k][j]
        B[k][j] = B[l][j]
        B[l][j] = temp

    return B

def RowScale(A,k,scale):
# =============================================================================
#     A is a NumPy array.  RowScale will return duplicate array with the
#     entries of row k multiplied by scale.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        B[k][j] *= scale

    return B

def RowAdd(A,k,l,scale):
# =============================================================================
#     A is a numpy array.  RowAdd will return duplicate array with row
#     l modifed.  The new values will be the old values of row l added to
#     the values of row k, multiplied by scale.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        B[l][j] += B[k][j]*scale

    return B
```

```python
# Tukar baris indeks ke-0 dengan baris indeks ke-1
A1 = RowSwap(A,0,1)
print(A1)
```

```python
# Tambahkan -8 kali baris indeks ke-0 ke baris indeks ke-1
A2 = RowAdd (A1,0,1,-8)
print(A2)
```

```python
# Tambahkan -2 kali baris indeks ke-0 ke baris indeks ke-2
A3 = RowAdd(A2,0,2,-2)
print(A3)
```

```python
# Kalikan baris indeks ke-1 dengan -1/7
A4 = RowScale(A3,1,-1/7)
print(A4)
```

```python
# Tambahkan 3 kali baris indeks ke-1 ke baris indeks ke-2
A5 = RowAdd(A4,1,2,3)
print(A5)
```

```python
# Kalikan baris indeks ke-2 dengan 1/4.85714286
A6 = RowScale(A5,2,1/4.85714286)
print(A6)
```

Persamaannya adalah :

$$
x_1 + 2x_2 + 3x_3 = 9\\
x_2 + 3.28571429x_3 = 7.57142857\\
x_3 = 2$$

**Substitusi Balik**

Selanjutnya, persamaan di-substitusi balik untuk menemukan nilai $x_1$ dan $x_2$ dengan nilai $x_3$ yang telah diketahui.

```python
X3 = 2
print('X3 =',X3)
X2 = 7.57142857 - (3.28571429 *2)
print('X2 =', round(X2))
X1 = 9 - 2*X2 - 3*X3
print('X1 =',round(X1))
```

## Ilustrasi 3D dengan Matplotlib

Matplotlib adalah modul python yang digunakan untuk menampilkan keluaran seperti grafik, plot, gambar, dan lain sebagainya. Kita dapat mengkombinasikan NumPy dan Matplotlib untuk mencetak grafik seperti di bawah ini.

```python
# solusi
X3 = 2
X2 = 1
X1 = 1
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm

# Persamaan 8x1 + 9x2 + x3 = 19
x1, y1, z1, w1 = 8, 9, 1, 19

# Persamaan x1 + 2x2 + 3x3 = 9
x2, y2, z2, w2 = 1, 2, 3, 9

# Persamaan 2x1 + x2 + x3 = 5
x3, y3, z3, w3 = 2, 1, 1, 5

x_axis, y_axis = np.linspace(0,20,10), np.linspace(0,20,10)
X, Y = np.meshgrid(x_axis, y_axis)
Z1 = (w1 - x1*X - y1*Y)/z1
Z2 = (w2 - x2*X - y2*Y)/z2
Z3 = (w3 + X - Y)/z3

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z1, alpha=1, cmap=cm.Accent, rstride=100, cstride=100)
ax.plot_surface(X, Y, Z2, alpha=1, cmap=cm.Paired, rstride=100, cstride=100)
ax.plot_surface(X, Y, Z3, alpha=1, cmap=cm.Pastel1, rstride=100, cstride=100)

ax.plot((X1,), (X2,), (X3,), lw=2, c='k', marker='o', markersize=7, markeredgecolor='g', markerfacecolor='white')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```
